```python
mask1=delivery['batter']=='V Kohli'
newdelivery=delivery[mask1]
newdelivery.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
def run_score(batsman):
  mask=delivery['batter']==batsman
  newdelivery=delivery[mask]
  return newdelivery.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
run_score('MS Dhoni')
```

```python
#isin()
#Find the most destructive death over  batsman in the history of IPL
#strike rate=(Number of runs/Number of balls)/100
#min batsman 200 balls in over 16-20
```

```python
mask2=delivery['over']>=16
mask2
death_over=delivery[mask2]
death_over.shape
```

```python

all_batsman=death_over.groupby('batter')['batsman_runs'].count()
x=all_batsman>200
batsman_list=all_batsman[x].index.tolist()
```

```python
#Runs scored by these 71 batsman
#Balls palyed by  these 71 batsman
#Strike rate of these 71 batsman

final=delivery[delivery['batter'].isin(batsman_list)]
```

```python
 run=final.groupby('batter')['batsman_runs'].sum()
```

```python
balls=final.groupby('batter')['ball'].count()
```

```python
sr=(run/balls)*100
```

```python
sr.groupby('batter').sum().sort_values(ascending=False).head(10)
```

```python
matches=pd.read_csv('/content/matches.csv')
matches.head()
```

```python
##Merge
#Orange Cap Holder
newdata=delivery.merge(matches,left_on='match_id',right_on='id')
newdata.shape
```

```python
newdata.groupby(['season','batter'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')[['season','batter']]
```

```python
#Pivot Table
matches.pivot_table(index='season',columns='toss_decision',values='id',aggfunc='count')
```

```python
delivery.pivot_table(index='over',columns='batsman_runs',values='ball',aggfunc='count')
```
