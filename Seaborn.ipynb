```python
#Seaborn
import seaborn as sns
```

```python
df=sns.load_dataset('tips')

```

```python
df.head()
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')

```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
df['smoker'].value_counts()
```

```python
sns.distplot(df['tip'],kde=True,bins=10)
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
#Categorical Plots

#countplot
sns.countplot(y='day',data=df)
```

```python
sns.barplot(x='sex',y='tip',data=df)
```

```python
sns.boxplot(y='sex',x='total_bill',data=df)

```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

```python
sns.violinplot(x='total_bill',y='day',data=df,hue='sex')
```
