```python
#Matplotlib

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph in 2D')
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*--')
```

```python
#Creating subplot
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
#Bar plot
X=[2,8,10]
Y=[11,16,9]

x2=[3,9,11]
y2=[12,17,10]

plt.bar(x2,y2)
```

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]

#box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
#pie chart
labels='Python','C++','Java','Ruby'
size=[215,130,245,230]
colors=['gold','yellowgreen','lightcoral','lightblue']
explode=(0.4,0,0,0)

plt.pie(size,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
